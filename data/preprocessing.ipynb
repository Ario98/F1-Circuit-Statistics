```python
import pandas as pd
import pprint
```

```python
# Read csv files
circuits = pd.read_csv('circuits.csv')
constructor_results = pd.read_csv('constructor_results.csv')
constructor_standings = pd.read_csv('constructor_standings.csv')
constructors = pd.read_csv('constructors.csv')
driver_standings = pd.read_csv('driver_standings.csv')
drivers = pd.read_csv('drivers.csv')
lap_times = pd.read_csv('lap_times.csv')
pit_stops = pd.read_csv('pit_stops.csv')
qualifying = pd.read_csv('qualifying.csv')
races = pd.read_csv('races.csv')
seasons = pd.read_csv('seasons.csv')
results = pd.read_csv('results.csv')
sprint_results = pd.read_csv('sprint_results.csv')
status = pd.read_csv('status.csv')
```

```python
# Join the necessary dataframes. In this case we are more circuit oriented.
df_merged = circuits.merge(races, on='circuitId')
df_merged = df_merged.merge(results, on='raceId')
df_merged = df_merged.merge(drivers, on='driverId')
df_merged = df_merged.merge(constructors, on='constructorId')

df_merged = df_merged.sort_values(by='circuitId', ascending=True)

#df_merged["full_name"] = df_merged["forename"] + df_merged["surname"]
df_merged['full_name'] = df_merged[['forename','surname']].apply(lambda x: ' '.join(x), axis=1)

df_merged.rename(columns = {'nationality_x':'nationality_driver', 'nationality_y':'nationality_constructor'},inplace = True)

df_merged = df_merged.reset_index(drop=True)

df_merged
```

```python
# export to csv
df_merged.to_csv('data.csv', header=True, index= False)
```
